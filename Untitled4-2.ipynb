```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms as T, models
from torch.cuda.amp import autocast, GradScaler
from PIL import Image
from scipy.io import loadmat
import csv

# Constants
IMG_SIZE = 224  # Vision Transformer expects 224x224 input size
BATCH_SIZE = 32
NUM_POINTS = 6  # 6 (x, y) pairs => 12 outputs
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")


# Dataset class
class DogHeartDataset(Dataset):
    def __init__(self, root, transforms=None, has_labels=True):
        self.root = root
        self.transforms = transforms
        self.has_labels = has_labels
        self.imgs = list(sorted(os.listdir(os.path.join(root, "Images"))))
        self.points = None
        if has_labels:
            self.points = list(sorted(os.listdir(os.path.join(root, "Labels"))))

    def __getitem__(self, idx):
        img_path = os.path.join(self.root, "Images", self.imgs[idx])
        img = Image.open(img_path).convert("RGB")
        w, h = img.size

        if self.transforms is not None:
            img = self.transforms(img)
        h_new, w_new = img.shape[1], img.shape[2]

        if self.has_labels:
            points_path = os.path.join(self.root, "Labels", self.points[idx])
            mat = loadmat(points_path)
            six_points = mat['six_points'].astype(float)
            six_points = torch.as_tensor(six_points, dtype=torch.float32)
            six_points[:, 0] = w_new / w * six_points[:, 0]
            six_points[:, 1] = h_new / h * six_points[:, 1]
            six_points = torch.reshape(six_points, (-1,)) / h_new  # Normalize points
            return img, six_points
        else:
            return img, self.imgs[idx]

    def __len__(self):
        return len(self.imgs)


# Data preprocessing
def get_transform(resized_image_size):
    return T.Compose([
        T.Resize((resized_image_size, resized_image_size)),
        T.ToTensor(),
        T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])


# Model class using Vision Transformer
class ViTRegressor(nn.Module):
    def __init__(self):
        super(ViTRegressor, self).__init__()
        self.base_model = models.vit_b_16(pretrained=True)  # Use ViT-B/16 from torchvision

        # Modify the classifier to output NUM_POINTS * 2 (12 outputs)
        in_features = self.base_model.heads.head.in_features
        self.base_model.heads.head = nn.Sequential(
            nn.Linear(in_features, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, NUM_POINTS * 2)
        )

        # Initialize weights
        self.base_model.heads.head.apply(self.initialize_weights)

    @staticmethod
    def initialize_weights(m):
        if isinstance(m, nn.Linear):
            nn.init.xavier_uniform_(m.weight)
            nn.init.constant_(m.bias, 0)

    def forward(self, x):
        return self.base_model(x)


# Function to calculate VHS score
def calculate_vhs(predictions):
    points = predictions.view(-1, NUM_POINTS, 2)
    AB = torch.sqrt((points[:, 1, 0] - points[:, 0, 0]) ** 2 + (points[:, 1, 1] - points[:, 0, 1]) ** 2)
    CD = torch.sqrt((points[:, 3, 0] - points[:, 2, 0]) ** 2 + (points[:, 3, 1] - points[:, 2, 1]) ** 2)
    EF = torch.sqrt((points[:, 5, 0] - points[:, 4, 0]) ** 2 + (points[:, 5, 1] - points[:, 4, 1]) ** 2)
    vhs = 6 * (AB + CD) / EF
    return vhs


# Save checkpoint
def save_checkpoint(model, optimizer, epoch, filepath="checkpoint.pth"):
    checkpoint = {
        "epoch": epoch,
        "model_state_dict": model.state_dict(),
        "optimizer_state_dict": optimizer.state_dict(),
    }
    torch.save(checkpoint, filepath)
    print(f"Checkpoint saved to {filepath}")


# Training function
def train_model(model, train_loader, val_loader, optimizer, scaler, epochs=10, checkpoint_path="checkpoint.pth"):
    criterion1 = nn.SmoothL1Loss()
    criterion2 = nn.MSELoss()
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, verbose=True)

    training_losses = []
    validation_losses = []

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        for batch_idx, (images, labels) in enumerate(train_loader):
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            optimizer.zero_grad()

            # Mixed precision
            with autocast():
                outputs = model(images)
                loss = criterion1(outputs, labels) + 0.1 * criterion2(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item()

        train_loss = running_loss / len(train_loader)
        training_losses.append(train_loss)
        print(f"Epoch {epoch + 1}/{epochs} | Training Loss: {train_loss:.4f}")

        # Validation loop
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(DEVICE), labels.to(DEVICE)
                outputs = model(images)
                loss = criterion1(outputs, labels) + 0.1 * criterion2(outputs, labels)
                val_loss += loss.item()

        val_loss = val_loss / len(val_loader)
        validation_losses.append(val_loss)
        print(f"Validation Loss: {val_loss:.4f}")
        scheduler.step(val_loss)

        # Free GPU memory
        torch.cuda.empty_cache()

    # Save checkpoint after training
    save_checkpoint(model, optimizer, epochs, filepath=checkpoint_path)

    return training_losses, validation_losses


# Testing function
def test_model(model, test_loader, output_csv="test_predictions.csv"):
    model.eval()  # Set model to evaluation mode
    predictions = []

    print("\nStarting Testing...")

    with torch.no_grad():
        for images, file_names in test_loader:
            images = images.to(DEVICE)

            # Mixed precision inference
            with autocast():
                outputs = model(images)
                vhs_score = calculate_vhs(outputs).cpu().item()

            for file_name in file_names:
                predictions.append({"file_name": file_name, "vhs_score": vhs_score})

    # Save predictions to a CSV file
    with open(output_csv, mode="w", newline="") as file:
        writer = csv.writer(file)
        #writer.writerow(["file_name", "vhs_score"])
        for prediction in predictions:
            writer.writerow([prediction["file_name"], prediction["vhs_score"]])

    print(f"Test predictions saved to {output_csv}")


# Paths
train_folder = "/content/Train/Train"
valid_folder = "/content/Valid/Valid"
test_folder = "/content/Images"
output_csv = "/content/test_predictions.csv"
checkpoint_path = "/content/Model_Checkpoint/best_model_checkpoint.pth"

# Load datasets
train_dataset = DogHeartDataset(train_folder, get_transform(IMG_SIZE), has_labels=True)
valid_dataset = DogHeartDataset(valid_folder, get_transform(IMG_SIZE), has_labels=True)
test_dataset = DogHeartDataset(test_folder, get_transform(IMG_SIZE), has_labels=False)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)  # Batch size 1 for testing

# Initialize model, optimizer, and scaler
model = ViTRegressor().to(DEVICE)
optimizer = optim.AdamW(model.parameters(), lr=1e-5, weight_decay=1e-5)
scaler = GradScaler()

if __name__ == "__main__":
    # Train the model
    num_epochs = 50
    train_model(model, train_loader, valid_loader, optimizer, scaler, epochs=num_epochs, checkpoint_path=checkpoint_path)

    # Load the best model for testing
    checkpoint = torch.load(checkpoint_path)
    model.load_state_dict(checkpoint["model_state_dict"])

    # Test the model
    test_model(model, test_loader, output_csv=output_csv)
```

```python
import matplotlib.pyplot as plt
from PIL import ImageDraw, Image


def visualize_predictions_and_ground_truth(model, dataset, image_names):
    """
    Display images with original (ground truth) and predicted points and lines (AB, CD, EF).
    Original lines are red, predicted lines are green.
    """
    model.eval()  # Set the model to evaluation mode

    for image_name in image_names:
        idx = dataset.imgs.index(image_name)  # Find the index of the image
        img, ground_truth_points = dataset[idx]  # Get the image and ground truth points

        # Add batch dimension and move the image to the device
        img_tensor = img.unsqueeze(0).to(DEVICE)

        with torch.no_grad():
            predicted_points = model(img_tensor).cpu().numpy().reshape(-1, 2)

        # Denormalize ground truth points
        img_path = os.path.join(dataset.root, "Images", image_name)
        original_img = Image.open(img_path).convert("RGB")
        w, h = original_img.size
        ground_truth_points = ground_truth_points.numpy().reshape(-1, 2)
        ground_truth_points[:, 0] *= w  # Denormalize x-coordinates
        ground_truth_points[:, 1] *= h  # Denormalize y-coordinates

        # Denormalize predicted points
        predicted_points[:, 0] *= w
        predicted_points[:, 1] *= h

        # Print the points
        print(f"Image: {image_name}")
        print("Original Points (Ground Truth):")
        print(ground_truth_points)
        print("Predicted Points:")
        print(predicted_points)

        # Draw the lines
        draw = ImageDraw.Draw(original_img)

        # Ground truth lines (red)
        for i, j in [(0, 1), (2, 3), (4, 5)]:
            draw.line(
                (ground_truth_points[i][0], ground_truth_points[i][1],
                 ground_truth_points[j][0], ground_truth_points[j][1]),
                fill="red",
                width=3
            )

        # Predicted lines (green)
        for i, j in [(0, 1), (2, 3), (4, 5)]:
            draw.line(
                (predicted_points[i][0], predicted_points[i][1],
                 predicted_points[j][0], predicted_points[j][1]),
                fill="green",
                width=3
            )

        # Display the image
        plt.figure(figsize=(8, 8))
        plt.imshow(original_img)
        plt.axis("off")
        plt.title(f"Image: {image_name}\nRed: Ground Truth, Green: Prediction")
        plt.show()


# Paths
valid_folder = "/content/Valid/Valid"

# Load validation dataset
valid_dataset = DogHeartDataset(valid_folder, get_transform(IMG_SIZE), has_labels=True)

# List of images to visualize
image_names = ["1420.png", "1479.png", "1530.png"]

# Visualize predictions and ground truth
visualize_predictions_and_ground_truth(model, valid_dataset, image_names)
```
